```python
import os
import pickle
import numpy as np
from collections import Counter

root = "../../workdir"

print("Files in workdir:")
print(os.listdir(root))
```

```python
train_labels = pickle.load(open(os.path.join(root, "train_label.pkl"), "rb"))
eval_labels  = pickle.load(open(os.path.join(root, "eval_label.pkl"), "rb"))

train_ids = [lbl for lbl, name in train_labels]
eval_ids  = [lbl for lbl, name in eval_labels]

print("TRAIN distribution:")
print(Counter(train_ids))

print("\nEVAL distribution:")
print(Counter(eval_ids))

print("\nUnique train classes:", sorted(set(train_ids)))
print("Unique eval classes :", sorted(set(eval_ids)))
```

```python
train_pose = np.load(os.path.join(root, "train_data.npy"))
train_obj  = np.load(os.path.join(root, "train_object_data.npy"))

print("TRAIN pose:", train_pose.shape)    # (N_train, T=48, M=4, V=17, C=3)
print("TRAIN obj :", train_obj.shape)     # (N_train, T=48, O=?, C=3)

eval_pose  = np.load(os.path.join(root, "eval_data.npy"))
eval_obj   = np.load(os.path.join(root, "eval_object_data.npy"))

print("EVAL pose :", eval_pose.shape)
print("EVAL obj  :", eval_obj.shape)
```

```python
train_names = [name for _, name in train_labels]
eval_names  = [name for _, name in eval_labels]

print("Any cam4 in TRAIN? ", any("cam4" in n for n in train_names))
print("All cams in EVAL are cam4? ", all("cam4" in n for n in eval_names))
```
